```python
# Import the necessary modules.
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define parameters
offset = 1e-4 # to avoid blowing up logs
phi = np.linspace(0 + offset, 1 - offset, 200)
chi = [0.1, 0.5, 1, 1.5, 2, 3, 4]

# set up plots
_, ax = plt.subplots(figsize=(5, 5))
# loop over chi
for x in chi:
    # calculate the free energy
    free_energy = x*phi*(1 - phi) + phi * np.log(phi) + (1-phi) * np.log(1-phi)
    # add to the plot
    ax.plot(phi, free_energy, label='$\chi = $' + str(x))

ax.set_xlabel('$\phi$')
ax.set_ylabel('free energy per site')
ax.legend()
```
